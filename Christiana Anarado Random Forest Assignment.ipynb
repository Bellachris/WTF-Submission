# IRIS DATASET

```python
#importing all the required libraries needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
#Uploading the in-built dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#To return all the function and methods of the object without the values
dir(iris)
```

```python
#loading the dataset
ird = pd.DataFrame(iris.data)
ird.head()
```

```python
ird["target"] = iris.target
```

```python
ird[0:15]
```

```python
#Training, modeling and predicition of the dataset
X = ird.drop("target", axis = "columns")
y = ird.target
```

```python
#Split the dataset into train and test model
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
#using random forest to fits a decision tree classifiers of the dataset and to predict the aaccuracy and control over-fitting
model = RandomForestClassifier (n_estimators = 20)
model.fit(X_train, y_train)
```

```python
y_test
```

```python
X_test
```

```python
#to automates the prediction of the dataset using X_test and comparing it with y_test
model.score(X_test, y_test)
```

```python
y_pred_value = model.predict(X_test)
```

```python
y_pred_value
```

##### CONFUSION MATRIX

```python
#To evaluate the accuracy of a classification.
con_max = confusion_matrix(y_test, y_pred_value)
con_max
```

```python
#To vicualize the confusion matrix using seaborn
plt.figure(figsize = (9,9))
sns.heatmap(con_max, annot = True, fmt=".3f", linewidths=.5, cmap="Blues_r")
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
```

# DIABETES DATASET

```python
#Uploading the in-built dataset
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
#To return all the function and methods of the object without the values
dir(diabetes)
```

```python
#loading the dataset
df = pd.DataFrame(diabetes.data)
df.head()
```

```python
df["target"] = diabetes.target
```

```python
df[0:10]
```

```python
#Training, modeling and predicition
X = df.drop("target", axis = "columns")
y = df.target
```

```python
#Split the dataset into train and test model
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
model = RandomForestClassifier (n_estimators = 30)
model.fit(X_train, y_train)
```

```python
y_test
```

```python
X_test
```

```python
model.score(X_test, y_test)
```

```python
y_predict_value = model.predict(X_test)
y_predict_value
```

##### CONFUSION MATRIX

```python
#To evaluate the accuracy of a classification.
confu_max = confusion_matrix(y_test, y_predict_value)
confu_max
```

```python
#To vicualize the confusion matrix using seaborn
plt.figure(figsize = (10,7))
sns.heatmap(confu_max, annot = True, fmt=".3f", linewidths=.5)
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
```
